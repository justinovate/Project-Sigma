## **DataFrames**
A DataFrame is a two-dimensional labeled data structure with columns of potentially 
different data types, similar to a spreadsheet or SQL table. 
It provides a powerful and flexible way to manipulate and analyze structured data in Python, 
offering functionalities for data analysis.

```python
# .DataFrame() is a constructor function used to create new DataFrame
# Creating an Empty DataFrame
df = pd.DataFrame()
df
```

```python
# Create a DataFrame using a list of lists
row_data = [["John", 30], ["Jane", 28], ["Smith", 26]]
df = pd.DataFrame(row_data, columns=["Name", "Age"])
df
```

```python
# Create a DataFrame using a dictionary of lists
data = {
    "Name": ["John", "Jane", "Smith"],
    "Age": [30, 28, 26]
}

df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame using a list of dictionaries
data = [
    {"Name": "John", "Age": 30},
    {"Name": "Jane", "Age": 28},
    {"Name": "Smith", "Age": 26}
]

df = pd.DataFrame(data)
df
```

```python
# You can use type() to check the datatype if it is Series or DataFrame
type(df)
```

It's similar to a one-column table or an array with associated labels, 
providing powerful indexing and manipulation capabilities in Python.

```python
series = pd.Series([1, 2, 3, 4, 5])
series
```

### **Pandas Data Types**

Numeric:
- Integer (int64): Represents whole numbers (e.g., 10, -5). 
    This is the default integer type in pandas. (64 bit integer)
- Float (float64): Represents numbers with decimals (e.g., 3.14, -12.5).
- Boolean (bool): Represents logical True or False values.
- Object: This is a versatile but less efficient type that can store various data types 
like strings, lists, or custom objects. 
    Pandas uses this type when it cannot infer a more specific data type.

```python
# Integer (int64)
int_series = pd.Series([1, 2, 3, 4, 5])
int_series
```

```python
# Float (float64)
float_series = pd.Series([3.14, -3.14, 0.0001, -0.0001])
float_series
```

```python
# Boolean (bool): (True = 1 or False = 0)
boolean_series = pd.Series([True, False, True, False])
boolean_series
```

```python
# Objects (Object / Mixed Data Types)
object_series = pd.Series([30, 3.14, True, "John"])
object_series
```

Specialized Data Types:
- Datetime (datetime64[ns]): Represents dates and times with nanosecond precision. 
    Useful for time-series data analysis.
- Timedelta (timedelta64[ns]): Represents durations between timestamps.
- Categorical: Represents categorical data with predefined categories. 
    Efficient for storing limited sets of categories.
- Sparse: Represents sparse data with many missing values. 
    Stores data efficiently by only keeping non-zero values.

```python
# DateTime (Timestamp / datetime64)
# N: to_datetime - conver string to datetime
datetime_series = pd.Series([
    pd.to_datetime("2024-08-27 07:00:00"),
    pd.to_datetime("2024-08-27 08:30:00"),
    pd.to_datetime("2024-08-27 10:00:00"),
    pd.to_datetime("11-09-2024 05:30:00", format="%d-%m-%Y %H:%M:%S"),
    pd.to_datetime("11-2024-09 06:30:59", format="%d-%Y-%m %H:%M:%S")
])
datetime_series
```

```python
# Timedelta
timedelta_series = pd.Series([
    pd.Timedelta(days=8, hours=3, minutes=15),
    pd.Timedelta(days=4, hours=3, minutes=15),
    pd.Timedelta(days=1, hours=3, minutes=15)
])
timedelta_series
```

```python
# Categorical
categorical_series = pd.Series(
    pd.Categorical(["Marketing", "Sales", "Operations", "IT", "Finance", "HR"])
)
categorical_series
```

```python
# Sparse - missing/null values inside of the series
# N: Can include null/missing values of that datatype
sparse_series = pd.Series(
    pd.arrays.SparseArray([30, 31, 32, pd.NA, 29, 42, pd.NA])
)
sparse_series
# NaN - Not a Number
```

### **Changing Data Types**

```python
# Step 1: Check datatype
int_series.dtype
```

```python
# Step 2: Change the datatype (float)

# astype() "as type": returns a new DataFrame where 
#           data is changed to the specified data type.

# Convert the integer to float
float_series = int_series.astype('float64')
float_series
```

```python
# Converting a float series to a string
string_series = float_series.astype('str')
string_series
```

```python
# Cannot convert directly from string to int if the contents of the string contains decimal point.
# Convert to float first, then convert to int

converted_to_float = string_series.astype('float64')
converted_to_float
```

```python
converted_to_int = converted_to_float.astype('int64')
converted_to_int
```

**Example: Sales Data Analysis**

You have a dataset of sales transactions that includes the product name, quantity sold, and sale price. 
You want to analyze the data to find the total revenue per product.

```python
data = {
    'Product Name':['Iced Tea', 'Hot Chocolate' , 'Lemonade', 'Coffee', 'Milkshake', 'Tea', 'Smoothie', 'Soda', 'Protein Shake', 'Matcha Latte'],
    'Type': ['Cold', 'Hot', 'Cold', 'Hot', 'Cold', 'Hot', 'Cold', 'Cold', 'Cold', 'Hot'],
    'Stock': [15, 15, 15, 15, 15, 15, 15, 15, 15, 15],
    'Quantity Sold':[6, 9, 13, 11, 8, 6, 14, 10, 8, 10],
    'Manufacturing Cost':[7, 10, 6, 8, 9, 7, 10, 11, 8, 9],
    'Market Price':[13, 20, 11, 15, 19, 14, 17, 18, 20, 12],
    'Rating': [1, 3, 5, 4, 3, 2, 5, 3, 2, 3]
}

sales_df = pd.DataFrame(data)
sales_df
```

```python
# Step 2: Get the Product Name column
sales_df['Product Name']
```

```python
# Step 3: Calculate the Total Revenue
sales_df['Total Revenue'] = sales_df['Quantity Sold'] * sales_df['Market Price']
sales_df
```

```python
# Gross Profit: difference between total revenue and the cost of goods (Cost per product multiply to the quantity sold).
sales_df['Gross Profit'] = sales_df['Total Revenue'] - (sales_df['Quantity Sold'] * sales_df['Manufacturing Cost'])
sales_df
```

## **Conditional Filtering**

```python
# Get all products that have total revenue greater than or equal 150
sales_df[sales_df["Total Revenue"] >= 150]
```

## **Apply**

The apply function in pandas is a powerful tool for working with DataFrames. 
It allows you to apply a custom function to each element (row or column) of the DataFrame 
and return a new DataFrame or Series based on the results.

```python
sales_df
```

```python
def discount(original_price):
    discount_rate = 0.10
    discount_amount = original_price  * discount_rate
    discounted_price = original_price - discount_amount
    return discounted_price

sales_df["10% Discounted Price"] = sales_df["Market Price"].apply(discount)
sales_df
```

```python
# Get the items with total revenue >= 150
def best_sellers(total_revenue):
    return total_revenue >= 150

df_best_sellers = sales_df[sales_df["Total Revenue"].apply(best_sellers)]
df_best_sellers
```

## Pandas Operators
Data Analysis:

- sum(): Calculates the sum of a Series or DataFrame
- mean(): Calculates the mean of a Series or DataFrame
- median(): Calculates the median of a Series or DataFrame
- std(): Calculates the standard deviation of a Series or DataFrame
- var(): Calculates the variance of a Series or DataFrame

Data Loading and Exploration:

- head(): Shows the first few rows of a DataFrame
- tail(): Shows the last few rows of a DataFrame
- info(): Displays information about the DataFrame, including data types and memory usage
- describe(): Generates summary statistics for each column (mean, standard deviation, etc.)

```python
print("Sum of Rating Column:", sales_df["Total Revenue"].sum())
```

```python
# Steps to get the mean/average
# 1. Add all the data.
# 2. Divide the data by how many data we have.
print("Mean/Average of Rating Column:", sales_df["Rating"].mean())
```

```python
# Steps to get the median
# 1. Sort the values
# 2. 2 Possible Options
#      1. Odd - You can get the middle value.
#      2. Even - Add the middle two and divide by 2.
rating_list = sales_df["Rating"].tolist() # reverse=True , descending order
rating_list.sort()
print(rating_list)
```

```python
print("Media of Rating Column:", sales_df["Rating"].median())
```

```python
# You can use this to sort the values.
sales_df["Rating"].sort_values()
```

```python
# Deviation: Distance of a unit from the mean.
# Standard Deviation: how far apart the values are from their mean (average).
# Data Analysis: Used to describe the distribution of data and identify outliers.
# LOW - Data is close to the mean. 
# HIGH - Data is distributed/spread out from the mean. (Presence of outliers.)
# Zero - All data points are identical.

print("Standard Deviation of Rating Column:", sales_df["Rating"].std().round(2))

# Outliers are data points that significantly deviate from the majority of the data in a dataset. 
# They can be either unusually high or unusually low compared to the rest of the data.

# .round(no_of_decimal_places): round your number to the nearest decimal point.
```

```python
# Variance: how far apart the values are from their mean (average).
# LOW - Data is close to the mean. 
# HIGH - Data is distributed/spread out from the mean. (Presence of outliers.)
# Zero - All data points are identical.

print("Variance of Rating Column", sales_df["Rating"].var().round(2))

# Difference between Standard Deviation and Variance
# Units: Variance is expressed in squared units of the original data, 
# while standard deviation is expressed in the same units as the original.
```

### **Aggregating Data** (.groupby)

Aggregating data involves summarizing data points into meaningful statistics, 
such as averages, sums, or counts, which can be achieved using GroupBy operations or pivot tables. 
This helps in understanding the dataset at a higher level.

```python
# Step 1: Get the unique values from the Type column.
sales_df["Type"].unique()
```

```python
# Step 2: Changing data type to category
sales_df["Type"] = sales_df["Type"].astype("category")
```

```python
# Step 3: Check the data type
sales_df["Type"].dtype
```

```python
sales_df.info()
```

```python
# Step 4
# Task: Get the total revenue of all the cold beverages combined and all the hot beverages combined.
total_revenue_based_on_type = sales_df.groupby('Type')['Total Revenue'].sum()
total_revenue_based_on_type
```

```python
# Step 5: Add the data inside the DataFrame
total_revenue_based_on_type_df = pd.DataFrame()
total_revenue_based_on_type_df['Total Revenue'] = sales_df.groupby('Type')['Total Revenue'].sum()
total_revenue_based_on_type_df
```

```python
# Step 6: Create a new column
total_revenue_based_on_type_df['Quantity Sold'] = sales_df.groupby('Type')['Quantity Sold'].sum()
total_revenue_based_on_type_df
```
