# **SQL METHODS**

In SQL, methods refer to functions or procedures that perform specific tasks within your queries. They are like pre-defined tools that you can use to manipulate, analyze, or interact with your data in various ways.

### **SELECT**

The SELECT statement in SQL retrieves data from one or more tables, allowing users to specify columns to fetch and apply filters, sorting, and aggregations to manipulate the data returned.

```python
cursor.execute("SELECT * FROM employees")
cursor.fetchall()
```

### **DISTINCT**

The **DISTINCT** keyword is used to return only distinct (unique) values in the result set of a query. It ensures that duplicate rows are removed from the output.

```python
cursor.execute("SELECT DISTINCT department FROM employees;")
cursor.fetchall()
```

```python
# If you have a tuple that contains one element, add a comma at the end
numbers = (1,)
print(numbers)
type(numbers)
```

### **LIMIT**

The **LIMIT** clause specifies the maximum number of rows to return in a query result.

```python
cursor.execute("""
SELECT * FROM employees
LIMIT 5;
""")
cursor.fetchall()
```

```python
# LIMIT starting_offset(not included), total number of records to displayy
cursor.execute("""
SELECT * FROM employees
LIMIT 3, 5;
""")
cursor.fetchall()
```

### **COUNT**

The **COUNT** function is used to count the number of rows in a table or the number of rows matching a specific condition.

```python
cursor.execute("SELECT COUNT(*) FROM employees;")
cursor.fetchall()
```

```python
cursor.execute("SELECT COUNT(DISTINCT department) FROM employees;")
cursor.fetchall()
```

### **WHERE**

The **WHERE** clause is used to filter data retrieved from a database based on specific conditions.

### **Comparison Operators:**
#### `=`: Equal to

```python
cursor.execute("SELECT * FROM employees WHERE department = 'IT';")
cursor.fetchall()
```

#### `!=` : Not equal to

```python
cursor.execute("SELECT * FROM employees WHERE department != 'IT';")
cursor.fetchall()
```

#### `<` : Less than

1. Employees that have poor performance

```python
cursor.execute("SELECT * FROM employees WHERE performance_rating < 3;")
cursor.fetchall()
```

#### `>` : Greater than

2. Employees that have great performance

```python
cursor.execute("SELECT * FROM employees WHERE performance_rating > 3;")
cursor.fetchall()
```

#### `<=` : Less than or equal to

```python
cursor.execute("SELECT * FROM employees WHERE salary <= 59000;")
cursor.fetchall()
```

#### `>=` : Greater than or equal to

```python
cursor.execute("SELECT * FROM employees WHERE salary >= 65000;")
cursor.fetchall()
```

### **Logical Operators:**
#### `AND`: Used to combine multiple conditions where both must be true.

```python
cursor.execute("""
SELECT * FROM employees 
WHERE department = 'Sales' AND performance_rating >= 3;
""")
cursor.fetchall()
```

#### `OR`: Used to combine multiple conditions where at least one must be true.

```python
query = """
    SELECT employee_id, first_name, last_name, department
    FROM employees
    WHERE department = ?
        OR department = ?;
"""
cursor.execute(query, ('Operations', 'Marketing'))
cursor.fetchall()
```

#### `NOT`: Used to negate a condition.

```python
cursor.execute("""
SELECT *
FROM employees
WHERE NOT department = 'IT';
""")
cursor.fetchall()
```

### **Special Operators:**
#### `BETWEEN`: Checks if a value falls within a specified range.

```python
cursor.execute("""
SELECT employee_id, last_name, salary
FROM employees
WHERE salary BETWEEN 55000 AND 60000;
""")
cursor.fetchall()
```

#### `IN`: Checks if a value belongs to a set of values.

```python
cursor.execute("""
SELECT employee_id, first_name, last_name, department
FROM employees
WHERE department
    IN ('IT', 'Marketing', 'Operations');
""")
cursor.fetchall()
```

#### `LIKE`: Used for pattern matching with wildcards.

```python
# %string = ending substring

cursor.execute("""
SELECT first_name, last_name
FROM employees
WHERE last_name LIKE '%son';
""")
cursor.fetchall()
```

```python
# %string% = contains the substring
cursor.execute("""
SELECT * FROM employees
WHERE hire_date LIKE '%-04-%';
""")
cursor.fetchall()
```

### **ORDER BY**

The **ORDER BY** clause allows you to sort the results of your SELECT queries in either ascending or descending order. It lets you sort the retrieved data based on one or more columns, making it easier to analyze and interpret.

#### `ASC:` Ascending order (lowest to highest). (Default)

```python
cursor.execute("""
SELECT last_name, first_name
FROM employees
ORDER BY last_name ASC
""")
cursor.fetchall()
```

#### `DESC:` Descending order (highest to lowest).

```python
cursor.execute("""
SELECT employee_id, last_name, department, performance_rating, salary
FROM employees
ORDER BY salary DESC
""")
cursor.fetchall()
```

#### Summary Report:
1. Employees from the IT department has the highest salary range.
2. HR department have the lowest salary range.
3. Megan Hill and Barbara Harris from the IT department have high salary but poor performance.
4. Patricia Taylor from the HR department have great peformance but have the lowest salary.

### **GROUP BY**

The **GROUP BY** clause is used to organize and summarize data by grouping rows with the same values in one or more columns. This helps you analyze trends, patterns, and aggregate statistics within your data.

#### Grouping with `COUNT()`

```python
cursor.execute("SELECT DISTINCT department FROM employees")
cursor.fetchall()
```

```python
cursor.execute("""
SELECT department, COUNT(*) 
FROM employees
GROUP BY department
ORDER BY COUNT(*) DESC;
""")
cursor.fetchall()
```

#### Grouping with `SUM()`

```python
cursor.execute("""
SELECT SUM(salary)
FROM employees;
""")
cursor.fetchall()
```

```python
cursor.execute("""
SELECT department, SUM(salary) AS total_salary
FROM employees
GROUP BY department
ORDER BY total_salary DESC;
""")
cursor.fetchall()
```

#### Grouping with `AVG()`

```python
cursor.execute("""
SELECT department, AVG(salary) AS mean_salary
FROM employees
GROUP BY department
ORDER BY mean_salary DESC;
""")
cursor.fetchall()
```

#### Grouping with `MAX()` and `MIN()`

```python
cursor.execute("""
SELECT department, MIN(salary) as min_salary, MAX(salary)
FROM employees
GROUP BY department
ORDER BY min_salary;
""")
cursor.fetchall()
```
