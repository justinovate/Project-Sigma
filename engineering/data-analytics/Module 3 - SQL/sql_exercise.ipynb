## **Activity 1** 

Practice performing CRUD operations (Create, Read, Update, Delete) on a database table.

### **Perform the following tasks:**

```python
import sqlite3
```

1. Connect to the database `abc_corp.db` and create the cursor.

```python
conn = sqlite3.connect("abc_corp.db")
cursor = conn.cursor()
```

2. `Create` a new table named books, with the following columns:
  * book_id (primary Key)
  * title (Title of the book)
  * author (Author of the book)
  * genre (Genre of the book)
  * publication_year (Year of publication)
  * price (Price of the Book)

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title VARCHAR(100) NOT NULL,
    author VARCHAR(100) NOT NULL,
    genre VARCHAR(50) NOT NULL,
    publication_year INT NOT NULL,
    price DECIMAL(8, 2) NOT NULL
);
""")
conn.commit()
```

3. `Insert` at least 5 new books into the books table.

```python
books = [
    ("The Silent Forest", "Maria Lopez", "Mystery", "2018", 12.99),
    ("Beyond the Stars", "James Carter", "Science Fiction", "2021", 15.50),
    ("Love in Manila", "Ana Reyes", "Romance", "2019", 9.75),
    ("The Last Empire", "David Chen", "Fantasy", "2017", 18.20),
    ("Mind Over Matter", "Sophia Patel", "Self-Help", "2022", 14.00)
]

cursor.executemany("""
INSERT INTO books (title, author, genre, publication_year, price)
VALUES (?, ?, ?, ?, ?)
""", books)

conn.commit()
```

4. `Retrieve` all books from the books table.

```python
cursor.execute("SELECT * FROM books")
cursor.fetchall()
```

5. `Update` the price of a specific book.

```python
cursor.execute("""
UPDATE books
SET price = 13.99
WHERE book_id = 5
""")
conn.commit()
```

6. `Delete` a book from the table.

```python
cursor.execute("""
DELETE FROM books
WHERE book_id = 5
""")
conn.commit()
```
